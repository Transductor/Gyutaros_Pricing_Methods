```python
import math
import numpy as np
import matplotlib.pyplot as plt
import random
import time
```

```python
class distribution_from_sample:
    def __init__(self,data):
        self.data = data
        self.size = np.size(data)

    def plot_distribution(self,partition_size,width=0.3):
        sample = np.sort(self.data)
        step = (sample[self.size-1]-sample[0])/partition_size
        i=0
        k=0
        image = []
        for j in range(partition_size):
            while i<self.size and sample[i]<j*step+sample[0]:
                i += 1
            image.append(i-k)
            k = i
        plt.bar(np.arange(sample[0],sample[sample.size-1],step),image,width)
```

```python
data = distribution_from_sample([np.random.normal(0,0.1) for n in range(1000000)])
```

```python
data.plot_distribution(100,0.007)
```
